# amyloid

```r
# MMRM (lme) stats for Amyloid (main text)

new_table_amyloid <- read.csv("data/new_table_withCov.csv")
new_table_amyloid <- subset(new_table_amyloid, ad_category != "Stable")

new_table_amyloid$Treatment_Information <- factor(new_table_amyloid$Treatment_Information, levels = c("Placebo", "LY3314814-20mg", "LY3314814-50mg"))
new_table_amyloid$time_point <- factor(new_table_amyloid$time_point, levels = c("1", "52", "104"))


# Fit your linear mixed-effects model (LMM)
model <- lmer(Amyloid_Beta ~ Treatment_Information * factor(time_point) * ad_category + Disease.status.at.baseline + AGE.x + (1 | EID), data = new_table_amyloid)

# Perform ANOVA to obtain p-values for fixed effects, including interactions
anova_result <- anova(model)

# Display the ANOVA table
print(anova_result)

# Obtain LS Means for Treatment_Information
mmrm_lsmeans <- lsmeans(model, specs = c("Treatment_Information"), type = "response")

# Print LS Means
print(mmrm_lsmeans)

ls_means <- lsmeans(model, c("time_point", "Treatment_Information"))
contrast_ls <- contrast(ls_means, interaction = "pairwise", by = "Treatment_Information", combine = c(-1, 1, 0))
#summary(contrast_ls_rapid)
# Filter results for the change between 104 and 1 for "Slow" ad_category
contrast_ls_104_1 <- subset(contrast_ls, time_point_pairwise == "1 - 104")
print(contrast_ls_104_1)





# Filter data for "Slow" ad_category
am_amy_slow <- subset(new_table_amyloid, ad_category == "Slow")

# Filter data for "Rapid" ad_category
am_amy_rapid <- subset(new_table_amyloid, ad_category == "Rapid")

# Fit the mixed-effects model for "Slow"
model_slow <- lmer(Amyloid_Beta ~ Treatment_Information * time_point + Disease.status.at.baseline + AGE.x + (1 | EID), data = am_amy_slow)

ls_means_slow <- lsmeans(model_slow, c("time_point", "Treatment_Information"))
contrast_ls_slow <- contrast(ls_means_slow, interaction = "pairwise", by = "Treatment_Information", combine = c(-1, 1, 0))
#summary(contrast_ls_slow)

# Fit the mixed-effects model for "Rapid"
model_rapid <- lmer(Amyloid_Beta ~ Treatment_Information * time_point + Disease.status.at.baseline + AGE.x + (1 | EID), data = am_amy_rapid)

ls_means_rapid <- lsmeans(model_rapid, c("time_point", "Treatment_Information"))
contrast_ls_rapid <- contrast(ls_means_rapid, interaction = "pairwise", by = "Treatment_Information", combine = c(-1, 1, 0))
#summary(contrast_ls_rapid)
# Filter results for the change between 104 and 1 for "Slow" ad_category
contrast_ls_104_1_slow <- subset(contrast_ls_slow, time_point_pairwise == "1 - 104")

# Display the filtered results for "Slow"
print("Results for Slow:")
print(contrast_ls_104_1_slow)

# Filter results for the change between 104 and 1 for "Rapid" ad_category
contrast_ls_104_1_rapid <- subset(contrast_ls_rapid, time_point_pairwise == "1 - 104")

# Display the filtered results for "Rapid"
print("Results for Rapid:")
print(contrast_ls_104_1_rapid)
```
